```python
import os
import tensorflow as tf
import numpy as np
import math
```

```python
data_dir = '/home/shashank/Desktop/BootcampStudents/Data/Task_1/'
batch_size = 32
```

```python
data = np.load(os.path.join(data_dir,'ANN.npz'))
train_x,train_y = data["train_x"],data["train_y"]
val_x, val_y = data["val_x"], data["val_y"]
test_x ,test_y = data["test_x"],data["test_y"]
print train_x.shape,train_y.shape
```

```python
def next_batch(batch_number):
    tr_x_batch = train_x[batch_number*batch_size:(batch_number+1)*batch_size]
    tr_y_batch = train_y[batch_number*batch_size:(batch_number+1)*batch_size]
    
    return tr_x_batch,tr_y_batch
    
```

```python
#placeholders
x_pl = tf.placeholder(tf.float32,shape=[None,102400])
y_pl = tf.placeholder(tf.float32,shape=[None,5])
```

```python
#hidden layer variables W,B
W_1 = tf.Variable(tf.random_normal([102400,128],mean=0,stddev=1/np.sqrt(102400)))
B_1 = tf.Variable(tf.zeros([128]))
```

```python
#output of 1st hidden layer after activation
h_1 = tf.nn.relu(tf.matmul(x_pl,W_1)+ B_1)
```

```python
#Dropout_layer
keep_prob = tf.placeholder(tf.float32)
h_1_drop = tf.nn.dropout(h_1, keep_prob)
```

```python
#output layer variables W,B
W_2 = tf.Variable(tf.random_normal([128,5],mean=0,stddev=1/np.sqrt(102400)))
B_2 = tf.Variable(tf.zeros([5]))
```

```python
#logits
h_2 = tf.matmul(h_1_drop,W_2)+B_2
```

```python
## LOSS 
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_pl, logits=h_2))

train_step = tf.train.GradientDescentOptimizer(1e-4).minimize(cross_entropy)
```

```python
#Accuracy
correct_prediction = tf.equal(tf.argmax(h_2, 1), tf.argmax(y_pl, 1))

accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
max_epochs = 100

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for ex in range(max_epochs):
        i=0
        print "epoch_no :",ex
        print '------------------------'
        while True:
            
            tr_x_batch, tr_y_batch = next_batch(i)
            
            train_step.run(feed_dict={x_pl: tr_x_batch, y_pl: tr_y_batch, keep_prob: 0.5})
            
            
            i+=1
            
            if(tr_x_batch.shape[0]!=32):
                break
                
        train_accuracy = accuracy.eval(feed_dict={x_pl: train_x, y_pl: train_y, keep_prob: 1.0})
        print 'training accuracy %g' % train_accuracy
        
        test_accuracy = accuracy.eval(feed_dict = {x_pl: test_x, y_pl: test_y, keep_prob: 1.0})
        print 'testing accuracy %g'% test_accuracy
        
        print '------------------------'
    
```
