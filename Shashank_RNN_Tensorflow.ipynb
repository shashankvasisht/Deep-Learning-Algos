```python
import tensorflow as tf
import numpy as np
import os
```

```python
data_dir = '/home/shashank/Desktop/BootcampStudents/Data/Task_1/'
batch_size = 32
```

```python
data = np.load(os.path.join(data_dir,'LSTM.npz'))
train_x,train_y = data["train_x"],data['train_y']
val_x,val_y = data['val_x'],data['val_y']
test_x,test_y = data['test_x'],data['test_y']
print train_x.shape,' ',train_y.shape
```

```python
def next_batch(batch_number):
    tr_x_batch  = train_x[batch_number*batch_size:(batch_number+1)*batch_size]
    tr_y_batch = train_y[batch_number*batch_size:(batch_number+1)*batch_size]
    return tr_x_batch,tr_y_batch
```

```python
x_pl = tf.placeholder(tf.float32,shape=[None,50,2048])
y_pl = tf.placeholder(tf.float32,shape=[None,5])
```

```python
neur_rnn = 128
cell = tf.nn.rnn_cell.BasicRNNCell(neur_rnn)
outputs,_ = tf.nn.dynamic_rnn(cell,x_pl,dtype=tf.float32)
```

```python
outputs_flattened = tf.reshape(outputs,[-1,50*neur_rnn])
print outputs_flattened.shape
```

```python
def Weight_Variable(shape):
    initial = tf.random_normal(shape,mean=0,stddev=1/np.sqrt(50*neur_rnn))
    return tf.Variable(initial)
```

```python
def Bias_Variable(shape):
    initial = tf.constant(0.1,shape=shape)
    return tf.Variable(initial)
```

```python
W_1 = Weight_Variable([50*neur_rnn,128])
B_1 = Bias_Variable([128])
```

```python
h_1 = tf.nn.leaky_relu(tf.matmul(outputs_flattened,W_1)+B_1)
```

```python
W_2 = Weight_Variable([128,5])
B_2 = Bias_Variable([5])
```

```python
y_conv = tf.matmul(h_1,W_2)+B_2
```

```python
cross_entropy = tf.reduce_mean(
    tf.nn.softmax_cross_entropy_with_logits(labels=y_pl, logits=y_conv))

train_step = tf.train.GradientDescentOptimizer(1e-3).minimize(cross_entropy)
```

```python
correct_prediction = tf.equal(tf.argmax(y_conv, 1), tf.argmax(y_pl, 1))

accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
max_epochs = 100
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for ex in range(max_epochs):
        i=0
        print "epoch_no :",ex
        print '------------------------'
        while True:
            tr_x_batch,tr_y_batch = next_batch(i)
            train_step.run(feed_dict={x_pl:tr_x_batch,y_pl:tr_y_batch})
            
            if (tr_x_batch.shape[0]!=32):
                break
            i += 1
        train_accuracy = accuracy.eval(feed_dict={x_pl:train_x,y_pl:train_y})
        print ' training accuracy:', train_accuracy
        
        test_accuracy = accuracy.eval(feed_dict={x_pl:test_x,y_pl:test_y})
        print ' test accuracy:',test_accuracy
        print'----------------------------'
#         while True:
#             tr_x_batch,tr_y_batch = next_batch(i)
#             test_accuracy += accuracy.eval(feed_dict={x_pl:tr_x_batch,y_pl:tr_y_batch})
#             if (tr_x_batch.shape[0]!=32):
#                 break
#         test_accuracy /= tr_x_batch.shape[0]
#         print ' test accuracy:', test_accuracy
        
        
        #test_accuracy = accuracy.eval(feed_dict={x_pl:test_x,y_pl:test_y})
```
